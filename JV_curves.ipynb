```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import MultipleLocator
```

```python
df = pd.read_excel('JV_EQE_BD_data/all_ETLs_JV_curves.xlsx')
df
```

```python

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(
    df['voltage'], df['cds_current'],
    linewidth=2, color='blue', marker='s', markeredgecolor='white',markersize=8,
    label='CdS', zorder=0,
)
ax.plot(
    df['voltage'], df['gas_current'],
    linewidth=2, color='red', marker='v', markeredgecolor='white',markersize=8,
    label='GaS',zorder=0,
)
ax.plot(
    df['voltage'], df['znse_current'],
    linewidth=2, color='green', marker='^', markeredgecolor='white',markersize=8,
    label='ZnSe',zorder=0,
)
ax.plot(
    df['voltage'], df['mgs_current'],
    linewidth=2, color='purple', marker='*', markeredgecolor='white',markersize=12,
    label='MgS',zorder=0,
)
ax.plot(
    df['voltage'], df['sro2_current'],
    linewidth=2, color='orange', marker='o', markeredgecolor='white',markersize=8,
    label='SrO2',zorder=0,
)

ax.legend()
ax.set_xlabel('Voltaje (V)')
ax.set_ylabel('Densidad de corriente (mA/cm\u00b2)')

fig.savefig('figures/JV_curves.svg')

'''#Inset real interval
xmin = 0.618; xmax = 0.622
ymin = 33; ymax = 34.75'''

'''#Inset box interval
width = 0.1
height = 5
center = ((xmax+xmin)/2, (ymax+ymin)/2)
xy = (center[0] - (width/2), center[1]-(height/2))

rect = matplotlib.patches.Rectangle(
    xy=xy,
    width=width,
    height=height,
    facecolor=(0,0,0,0),
    edgecolor='blue',
    linestyle='--',
    zorder=1
)


ax.add_artist(rect)'''
```
